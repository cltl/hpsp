## This notebook study the overlap of OntoNotes and the test data to establish an upperbound.

```python
# the file is generated by notebook/extracting-triples.ipynb
triple_path = '../output/ontonote-svo.2018-06-11-d63ddf1.txt'
# dataset from Wang et al.
pos_path = '../data/Wang_et_al_NAACL18/data/pos-all.txt'
neg_path = '../data/Wang_et_al_NAACL18/data/neg-all.txt'
```

```python
import pandas as pd
```

```python
col_names = ['subj', 'verb', 'dobj']
```

```python
df = pd.read_csv(triple_path, sep='\t', header=0, names=col_names)
df.head()
```

```python
len(df)
```

```python
pos = pd.read_csv(pos_path, sep=' ', header=0, names=col_names)
pos.head()
```

```python
len(pos)
```

```python
neg = pd.read_csv(neg_path, sep=' ', header=0, names=col_names)
neg.head()
```

```python
len(neg)
```

## Verb coverage

```python
pos_verbs = pos[['verb']].drop_duplicates()
onto_verbs = df[['verb']].drop_duplicates()
overlap_verbs = pd.merge(pos_verbs, onto_verbs, on='verb')
```

```python
len(overlap_verbs) / len(pos_verbs)
```

## Subject coverage

```python
pos_subjs = pos[['subj']].drop_duplicates()
onto_subjs = df[['subj']].drop_duplicates()
overlap_subjs = pd.merge(pos_subjs, onto_subjs, on='subj')
```

```python
len(overlap_subjs) / len(pos_subjs)
```

## Object coverage

```python
pos_dobjs = pos[['dobj']].drop_duplicates()
onto_dobjs = df[['dobj']].drop_duplicates()
overlap_dobjs = pd.merge(pos_dobjs, onto_dobjs, on='dobj')
```

```python
len(overlap_dobjs) / len(pos_dobjs)
```

## Subject-verb, object-verb, and full overlap coverage

```python
len(pd.merge(pos[['subj', 'verb']].drop_duplicates(), df, on=['subj', 'verb']))
```

```python
len(pd.merge(pos[['dobj', 'verb']].drop_duplicates(), df, on=['dobj', 'verb']))
```

```python
len(pd.merge(pos, df, on=['subj', 'verb', 'dobj']))
```
